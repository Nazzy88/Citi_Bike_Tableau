```python
# Dependencies & Setup
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Files to Load
july2013 = "./July/201307-citibike-tripdata.csv"
july2014 = "./July/201407-citibike-tripdata.csv"
july2015 = "./July/201507-citibike-tripdata.csv"
july2016 = "./July/201607-citibike-tripdata.csv"
july2017 = "./July/201707-citibike-tripdata.csv"
july2018 = "./July/201807-citibike-tripdata.csv"
july2019 = "./July/201907-citibike-tripdata.csv"
july2020 = "./July/202007-citibike-tripdata.csv"

# Read July Citi Bike Data Files & Store into Pandas DataFrames
july2013_df = pd.read_csv(july2013, encoding="ISO-8859-1")
july2014_df = pd.read_csv(july2014, encoding="ISO-8859-1")
july2015_df = pd.read_csv(july2015, encoding="ISO-8859-1")
july2016_df = pd.read_csv(july2016, encoding="ISO-8859-1")
july2017_df = pd.read_csv(july2017, encoding="ISO-8859-1")
july2018_df = pd.read_csv(july2018, encoding="ISO-8859-1")
july2019_df = pd.read_csv(july2019, encoding="ISO-8859-1")
july2020_df = pd.read_csv(july2020, encoding="ISO-8859-1")
```

```python
# Combine All July DataFrames from 2013-2020 Into Single DataFrame
combined_july_df = july2013_df.append([july2014_df, july2015_df, july2016_df, july2017_df, july2018_df, july2019_df, july2020_df], ignore_index=True)
combined_july_df.head()
```

```python
# Rename Columns
combined_july_df = combined_july_df.rename(columns={"tripduration": "Trip Duration",
                                            "starttime": "Start Time",
                                            "stoptime": "Stop Time",
                                            "start station id": "Start Station ID",
                                            "start station name": "Start Station Name",
                                            "start station latitude": "Start Station Latitude",
                                            "start station longitude": "Start Station Longitude",
                                            "end station id": "End Station ID",
                                            "end station name": "End Station Name",
                                            "end station latitude": "End Station Latitude",
                                            "end station longitude": "End Station Longitude",
                                            "bikeid": "Bike ID",
                                            "usertype": "User Type",
                                            "birth year": "Birth Year",
                                            "gender": "Gender"
                                           })
combined_july_df.head()
```

```python
# Convert All DateTimes to "%Y-%m-%d %H:%M:%S" Format
combined_july_df["Start Time"] = pd.to_datetime(combined_july_df["Start Time"])
combined_july_df["Stop Time"] = pd.to_datetime(combined_july_df["Stop Time"])
combined_july_df.head()
```

```python
# Replace Gender Codes with values from the Citi Website
combined_july_df['Gender'].replace(to_replace=[0, 1, 2],
           value=['Unknown', 'Male', 'Female'],
           inplace=True)
```

```python
# Save Combined DataFrames to CSV
combined_july_df.to_csv("july-citibike-tripdata.csv", index=False, header=True)
```
